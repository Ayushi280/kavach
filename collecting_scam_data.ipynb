```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("M-Arjun/SpamShield-Datasets")

print("SPLITS + COLUMNS:")
print(ds)
```

```python
import pandas as pd
df = ds["train"].to_pandas()

print("LABEL values:")
print(df["label"].value_counts())
print("\nCATEGORY values:")
print(df["category"].value_counts())
print("\nLANGUAGE values (top 10):")
print(df["language"].value_counts().head(10))
print("\nSAMPLE ROWS:")
print(df.sample(5)[["text","label","category","language"]].to_string())
```

```python
out = pd.DataFrame({
    "text": df["text"].astype(str).str.strip(),
    "label": df["label"].map({1: "scam", 0: "safe"}),
    "category": df["category"],
    "language": "en",
    "type": "sms"
})

# clean up
out = out[out["text"].str.len() > 3]        # drop empties/junk
out = out.drop_duplicates(subset=["text"])  # drop dupes

out.to_csv("spamshield_clean.csv", index=False)

print(out["label"].value_counts())
print("Total rows:", len(out))
print(out.head())
```

```python
import re
def is_latin_text(s):
  s=str(s)
  if len(s)<4:
    return False
  letters=re.findall(r'[a-zA-Z]', s)
  return len(letters)/len(s)>0.6

clean=out[out["text"].apply(is_latin_text)].copy()
print("Before:", len(out), " After:", len(clean))
print(clean["label"].value_counts())

clean.to_csv("spamshield_clean.csv", index=False)
print(clean.head(10)[["text","label"]].to_string())


```

```python

def has_foreign_script(s):
    return bool(re.search(r'[\u0600-\u06FF\u0900-\u097F\u4e00-\u9fff\uAC00-\uD7AF\u0400-\u04FF]', str(s)))

clean2 = clean[~clean["text"].apply(has_foreign_script)].copy()
print("After removing foreign scripts:", len(clean2))
clean2.to_csv("spamshield_clean.csv", index=False)
```
